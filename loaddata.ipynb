```python
import pandas as pd
```

Load all the dataset to integrate 

```python
spleen = pd.read_csv("Spleen-counts.csv")
thymus = pd.read_csv("Thymus-counts.csv")
microglia = pd.read_csv("Brain_Microglia-counts.csv")
neuron = pd.read_csv("Brain_Neurons-counts.csv")
marrow = pd.read_csv("Marrow-counts.csv")
metadata = pd.read_csv("metadata_FACS.csv")
ann = pd.read_csv('annotations_FACS.csv')
```

Add a labels column to all the datasets

```python
spleen.rename(columns = {'Unnamed: 0':'Genes'}, inplace = True)
thymus.rename(columns = {'Unnamed: 0':'Genes'}, inplace = True)
microglia.rename(columns = {'Unnamed: 0':'Genes'}, inplace = True)
neuron.rename(columns = {'Unnamed: 0':'Genes'}, inplace = True)
marrow.rename(columns = {'Unnamed: 0':'Genes'}, inplace = True)
```

Set Genes as index

```python
spleen.set_index('Genes', inplace=True)
thymus.set_index('Genes', inplace=True)
microglia.set_index('Genes', inplace=True)
neuron.set_index('Genes', inplace=True)
marrow.set_index('Genes', inplace=True)
```

```python
# Concatenate all the count files on x-axis
complete_data = pd.concat([spleen, thymus, microglia, neuron, marrow], axis = 1)
```

```python
complete_data_T = complete_data.T
```
